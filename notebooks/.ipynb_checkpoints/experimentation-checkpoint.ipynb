```python
import pandas as pd
from sklearn.model_selection import train_test_split
import sys
from pathlib import Path

# Add the root directory to sys.path (one level up from notebooks/)
root_dir = Path().resolve().parent
sys.path.append(str(root_dir))

# Load dataset
df = pd.read_csv("../data/raw/YearPredictionMSD.txt", header=None)

# Add column names: first column is the year, rest are audio features
df.columns = ['year'] + [f'feature_{i}' for i in range(1, 91)]

# Convert year to decade (e.g., 2001 â†’ 2000s)
df['decade'] = (df['year'] // 10 * 10).astype(int)

# Print how many samples are in each decade
print("Samples per decade:")
print(df['decade'].value_counts().sort_index())

# Split into train (70%), temp (30%)
train_df, temp_df = train_test_split(df, test_size=0.3, stratify=df['decade'], random_state=42)

# Split temp into validation (15%) and test (15%)
val_df, test_df = train_test_split(temp_df, test_size=0.5, stratify=temp_df['decade'], random_state=42)

# Confirm shapes
print(f"Train shape: {train_df.shape}")
print(f"Validation shape: {val_df.shape}")
print(f"Test shape: {test_df.shape}")
```

```python
from src.collections_net import *
from torch.utils.data import DataLoader, TensorDataset
import matplotlib.pyplot as plt
import numpy as np

# Prepare the data (X = features, y = decade)
X_train = torch.tensor(train_df.iloc[:, 1:-1].values, dtype=torch.float32)
y_train = torch.tensor(train_df['decade'].astype('category').cat.codes.values, dtype=torch.long)

train_dataset = TensorDataset(X_train, y_train)
train_loader = DataLoader(train_dataset, batch_size=128, shuffle=True)
```

```python
# Dictionary to store loss trajectories
loss_dict = {}

# List of models
model_classes = [ModelA, ModelB, ModelC, ModelD, ModelE, ModelF, ModelG, ModelH, ModelI, ModelJ]
model_names = ['ModelA', 'ModelB', 'ModelC', 'ModelD', 'ModelE', 'ModelF', 'ModelG', 'ModelH', 'ModelI', 'ModelJ']
```

```python
# Training Loop 
for model_class, model_name in zip(model_classes, model_names):
    print(f"\n Training {model_name}")
    
    model = model_class()
    optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
    criterion = nn.CrossEntropyLoss()
    
    mini_batch_losses = []
    model.train()
    
    for batch_idx, (features, labels) in enumerate(train_loader):
        if batch_idx == 15:
            break
            
        optimizer.zero_grad()
        outputs = model(features)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        
        mini_batch_losses.append(loss.item())
        print(f"Mini-batch {batch_idx+1} | Loss: {loss.item():.4f}")
    
    loss_dict[model_name] = mini_batch_losses
```

```python
# Plotting mini-batch loss curves
plt.figure(figsize=(14, 8))
for model_name, losses in loss_dict.items():
    plt.plot(np.arange(1, 16), losses, label=model_name)

plt.xlabel('Mini-batch Number')
plt.ylabel('Loss')
plt.title('Mini-batch Loss Trajectories (First 15 batches)')
plt.legend()
plt.grid(True)
plt.savefig("C:/Users/Gharat/Downloads/ML-Assignment-3/images/01-model-comaparison-plots.png")
plt.show()
```
